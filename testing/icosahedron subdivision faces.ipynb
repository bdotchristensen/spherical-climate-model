```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances, euclidean_distances

from itertools import combinations
```

```python
phi = (1 + np.sqrt(5)) / 2
icosahedron = []

for k in range(3):
    
    for i in range(2):
        for j in range(2):
            if k == 0:
                t = (0, (-1)**j, phi * (-1)**i)
                icosahedron.append(t)

            elif k == 1:
                t = ((-1)**j, phi * (-1)**i, 0)
                icosahedron.append(t)

            elif k == 2:
                t = (phi * (-1)**i, 0, (-1)**j)
                icosahedron.append(t)

icosahedron = np.stack(icosahedron)
vertices = icosahedron / np.linalg.norm(icosahedron, axis = 1)[0]
```

```python
combos = list(combinations(range(12), 3))
allfaces = np.stack([np.mean(vertices[i,:], axis = 0) for i in combos])
```

```python
magnitudes = np.linalg.norm(allfaces, axis = 1)
faces_aux = allfaces[magnitudes > magnitudes.max() - 1e-6]
faces_aux.shape # keep in mind this isnt normalized yet to fit the unit sphere
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(faces_aux[:,0], faces_aux[:,1], faces_aux[:,2], s = 30)
```

```python
faces = faces_aux / np.linalg.norm(faces_aux, axis = 1).reshape(-1, 1)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(faces[:,0], faces[:,1], faces[:,2], s = 30)
```
